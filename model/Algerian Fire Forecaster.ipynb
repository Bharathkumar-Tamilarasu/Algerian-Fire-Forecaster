# **About the Dataset**

### Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class) The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

### Attribute Information:


* Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
* Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
* RH : Relative Humidity in %: 21 to 90
* Ws :Wind speed in km/h: 6 to 29
* Rain: total day in mm: 0 to 16.8 FWI Components
* Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
* Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
* Drought Code (DC) index from the FWI system: 7 to 220.4
* Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
* Buildup Index (BUI) index from the FWI system: 1.1 to 68
* Fire Weather Index (FWI) Index: 0 to 31.1
* Classes: two classes, namely Fire and not Fire

## **4. Model Building**

```python
# Independent and Dependent Variables
X=df_copy.drop(['FWI'],axis=1)
y=df_copy['FWI']
```

```python
# Train Test Split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=None)
```

```python
print('Shapes\nX Train ->',X_train.shape)
print('X Test ->',X_test.shape)
print('Y Train ->',y_train.shape)
print('Y Test ->',y_test.shape)
```

###i) Feature Selection Based on Correlation

```python
corr =  X_train.corr()
plt.figure(figsize=(12,10))
plt.title('Correlation between Independent Features')
sns.heatmap(corr,annot=True)
plt.show()
```

```python
# Checking columns with High Correlation i.e 0.85

def correlation_check(dataset,threshold):
  corr_matrix = dataset.corr()
  corr=set()
  col_list = dataset.columns
  for i in range(len(col_list)):
    for j in range(i+1,len(col_list)):
      if(corr_matrix.iloc[i,j]>threshold):
        corr.add(col_list[j])
  return corr
```

```python
corr_features = correlation_check(X_train,0.85)
corr_features
```

```python
# Dropping above columns in X_train and X_test

X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
print('Shapes\nX Train ->',X_train.shape)
print('X Test ->',X_test.shape)
```

###ii) Feature Scaling Or Standardization

---



```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

###iii) Linear Regression

```python
#Prediction

lr = LinearRegression()
lr.fit(X_train_scaled,y_train)
y_pred = lr.predict(X_test_scaled)

# Errors

mae = mean_absolute_error(y_test,y_pred)
score  = r2_score(y_test,y_pred)

print("Mean Absolute Error:",mae)
print("R2 Score:",score)
print('\n')

# Graphs

plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred)
plt.title('Y Test vs Y Predicted (Linear Regression)')
plt.xlabel('Y Test')
plt.ylabel('Y Predicted')
plt.show()
```

###iv) Lasso Regression

```python
#Prediction

ls = Lasso()
ls.fit(X_train_scaled,y_train)
y_pred = ls.predict(X_test_scaled)

# Errors

mae = mean_absolute_error(y_test,y_pred)
score  = r2_score(y_test,y_pred)

print("Mean Absolute Error:",mae)
print("R2 Score:",score)
print('\n')

# Graphs

plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred)
plt.title('Y Test vs Y Predicted (Lasso Regression)')
plt.xlabel('Y Test')
plt.ylabel('Y Predicted')
plt.show()
```

###v) Ridge Regression

```python
#Prediction

rg = Ridge()
rg.fit(X_train_scaled,y_train)
y_pred = rg.predict(X_test_scaled)

# Errors

mae = mean_absolute_error(y_test,y_pred)
score  = r2_score(y_test,y_pred)

print("Mean Absolute Error:",mae)
print("R2 Score:",score)
print('\n')

# Graphs

plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred)
plt.title('Y Test vs Y Predicted (Ridge Regression)')
plt.xlabel('Y Test')
plt.ylabel('Y Predicted')
plt.show()
```

###vi) ElasticNet Regression

```python
#Prediction

en = ElasticNet()
en.fit(X_train_scaled,y_train)
y_pred = en.predict(X_test_scaled)

# Errors

mae = mean_absolute_error(y_test,y_pred)
score  = r2_score(y_test,y_pred)

print("Mean Absolute Error:",mae)
print("R2 Score:",score)
print('\n')

# Graphs

plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred)
plt.title('Y Test vs Y Predicted (Ridge Regression)')
plt.xlabel('Y Test')
plt.ylabel('Y Predicted')
plt.show()
```

## **5. Prediction**

Leverage the trained models to make predictions on new data, assessing their performance and reliability.

As we can see above, the Linear Regression model provides the best score. Hence we will use this model for further predictions of FWI

```python
def values_predictor(temperature,rh,ws,rain,ffmc,dmc,isi,classes,region):
  temp_df = pd.DataFrame([[temperature,rh,ws,rain,ffmc,dmc,isi,classes,region]],columns=X_test.columns)
  scaled_input = scaler.transform(temp_df)
  predicted_value = lr.predict(scaled_input)
  return float(predicted_value)
```

The FWI has six levels of danger:

* **Very low danger** - Less than 5
* **Low danger** - Between 5 and 8.9
* **Moderate danger** - Between 9 and 16.4
* **High danger** - Between 16.5 and 30.4
* **Very high danger** - Between 30.5 and 42.4
* **Extreme danger** -  Greater than 42.4

```python
def danger_predictor(val):
  if val < 5:
    return 'very low danger'
  elif val >= 5 and val <= 8.9:
    return 'low danger'
  elif val >= 9 and val <= 16.4:
    return 'moderate danger'
  elif val >= 16.5 and val <= 30.4:
    return 'high danger'
  elif val >= 30.5 and val <= 42.4:
    return 'very high danger'
  else:
    return 'extreme danger'
```

```python
temperature = 35
rh = 48
ws = 18
rain = 0.0
ffmc = 90.1
dmc = 54.2
isi = 12.5
classes = 0
region = 0

predict = values_predictor(temperature,rh,ws,rain,ffmc,dmc,isi,classes,region)
predict = round(predict,1)
print(f'FWI is {predict} and the area is in {danger_predictor(predict)} of fire')
```
